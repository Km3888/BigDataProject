# Distribution of number of unique DOS IDs over time

Percentage change

```python
pctchange_2019 =pd.read_csv('pctchange_2019.csv')
```

```python
pctchange_2019.head()
```

```python
pctchange_2020 =pd.read_csv('pctchange_2020.csv')
```

```python
pctchange_2020.head()
```

```python
len(set(pctchange_2020['County']))
```

```python
len(set(pctchange_2019['County']))
```

```python
plt.figure(figsize = (32,32))
c =sns.color_palette("Spectral", as_cmap=True)
plt.title(" Percentage change of Active Corporations 2020")
ax = sns.heatmap(pctchange_2020.pivot("County", "month", "DOS ID"),  cmap=c, annot=True)
```

```python
plt.figure(figsize = (32,32))
c =sns.color_palette("Spectral", as_cmap=True)
plt.title(" Percentage change of Active Corporations 2020 After March")
ax = sns.heatmap(pctchange_2020[pctchange_2020['month']>=3].pivot("County", "month", "DOS ID"),  cmap=c, annot=True)
```

```python
plt.figure(figsize = (32,32))
c =sns.color_palette("Spectral", 5,as_cmap=True)
plt.title("Percentage change of Active Corporations 2019")
ax = sns.heatmap(pctchange_2019.pivot("County", "month", "DOS ID"),  cmap=c, annot=True)
```

```python
pct_mean_2019 = pctchange_2019.groupby(by=['County'])['DOS ID'].mean()
pct_mean_2019_df = pd.DataFrame(pct_mean_2019)
```

```python
plt.figure(figsize = (16,32))
c =sns.color_palette("Spectral", 5,as_cmap=True)
plt.title("Average percentage change of Active Corporations 2019")
ax = sns.heatmap(pct_mean_2019_df,  cmap=c, annot=True)
```

```python
pct_mean_2020 = pctchange_2020.groupby(by=['County'])['DOS ID'].mean()
pct_mean_2020_df = pd.DataFrame(pct_mean_2020)
```

```python
plt.figure(figsize = (16,32))
c =sns.color_palette("Spectral", 5,as_cmap=True)
plt.title("Average percentage change of Active Corporations 2020")
ax = sns.heatmap(pct_mean_2020_df,  cmap=c, annot=True)
```

```python
all_counties = list(pct_mean_2020_df.index)
```

```python
plt.figure(figsize = (16,32))
plt.title("Average percentage change of Active Corporations 2020")
g=sns.scatterplot('DOS ID', 'County', data=pct_mean_2020_df, hue=all_counties, s=256)
g.get_legend().set_visible(False)
plt.xlabel('Percentge change')
plt.show()
```

```python
pctchange_2020_afterMarch = pctchange_2020[pctchange_2020['month']>=3]
pct_mean_2020_afterMarch = pctchange_2020_afterMarch.groupby(by=['County'])['DOS ID'].mean()
pct_mean_2020_df_afterMarch = pd.DataFrame(pct_mean_2020_afterMarch)
```

```python
plt.figure(figsize = (16,16))
plt.title("Average percentage change of Active Corporations 2020 After March")
g=sns.scatterplot('DOS ID', 'County', data=pct_mean_2020_df_afterMarch, hue=all_counties, s=256)
g.get_legend().set_visible(False)
plt.xlabel('Percentge change')
plt.show()
```
